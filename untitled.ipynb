```python
import matplotlib.pyplot as plt
from model_diagnostics.calibration import compute_bias, plot_bias, plot_reliability_diagram
import numpy as np
import polars as pl


n_obs = 100
y_obs = np.arange(n_obs)
y_pred = pl.DataFrame({
    "model_1": np.arange(n_obs) + 10 * n_obs * np.sin(y_obs),
    "model_2": (y_obs - 5)**2,
    "model_3": (y_obs - 3)**2,
})

rng = np.random.default_rng(42)
# NaN and Null is not the same!
feature_missing = rng.integers(low=0, high=n_obs//2, size=n_obs).astype(float)
feature_missing[::5] = np.nan
feature_missing = pl.Series(feature_missing).set_at_idx([0, 1], None)
```

```python
cond = feature_missing.is_not_null() & feature_missing.is_not_nan()
plot_bias(
    y_obs=y_obs[cond],
    y_pred=y_pred.filter(cond),
    feature=feature_missing.filter(cond),
)
```

```python
ax = plot_bias(y_obs=y_obs, y_pred=y_pred, feature=feature_missing, with_errorbars=True)
```

```python
ax = plot_bias(y_obs=y_obs, y_pred=y_pred, feature=feature_missing, with_errorbars=False)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 8))

for i, ax in enumerate(axes.flat):
    plot_bias(y_obs=y_obs, y_pred=y_pred, feature=feature_missing, n_bins=i+1, with_errorbars=True, ax=ax)
    ax.set_title(f"Bias Plot i={i+1}")
    ax.get_legend().remove()

# legend of last axis
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc="outside right upper", bbox_to_anchor=(1.1, 1))

fig.set_tight_layout(True)
```

```python
import pandas as pd
import pyarrow as pa
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import calibration_curve, CalibrationDisplay


X, y = make_classification(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
```

```python
y_prob = clf.predict_proba(X_test)[:, 1]
prob_true, prob_pred = calibration_curve(y_test, y_prob, n_bins=10)
disp = CalibrationDisplay(prob_true, prob_pred, y_prob)
disp.plot()
```
